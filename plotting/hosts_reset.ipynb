```python
import numpy as np
import pandas as pd

# next command ensures that plots appear inside the notebook
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns  # also improves the look of plots
sns.set()  # set Seaborn defaults
plt.rcParams['figure.figsize'] = [10, 5]  # default hor./vert. size of plots, in inches
plt.rcParams['lines.markeredgewidth'] = 1  # to fix issue with seaborn box plots; needed after import seaborn
```

```python
df = pd.read_csv("../local_reset_100.csv")
```

```python
filter_col = [col for col in df if col.startswith('timings.')]
```

```python
df[filter_col]
```

```python
# plot the timings for each run
to_plot = ['timings.setup', 'timings.online']
df_plotting = df[to_plot].copy() / 1000 # for seconds
ax = df_plotting.plot()
ax.set_ylabel("Time (s)")
ax.set_xlabel("Run number")
ax.set_title("Duration of SFE, resetting the ABY party between each run")
ax.legend(['Setup', 'Online'])
```
